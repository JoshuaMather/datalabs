<h2>Cleaning and converting to an string data type to a number data type</h2>
<p>The percentage column is a string data type because of the % character.  Remove the % character and convert the values for this column to a numeric data type</p>

```python
df
for i in range(0,len(df.index)):
    value = df.at[i, 'Percentage']
    print(value)
    if type(value) != int:
        newvalue = value.replace('%', '')

        newvalue = int(newvalue)
        print(newvalue)

        df = df.replace(to_replace = value, value = newvalue)
    
```

```python
df
df.replace({'Percentage': "26%"}, 26)
```

<h2>New Dataframes</h2>
<p>Often it is useful to create seperate smaller dataframes, create four new dataframes named df_first, df_upp_sec, df_low_sec and df_third, to seperate the different degree classifications.</p>

Example:

```python
mask = df['Classification of First Degree'] == "First"
df_first = df[mask]
df_first

mask = df['Classification of First Degree'] == "Upper second"
df_upp_sec = df[mask]
df_upp_sec

mask = df['Classification of First Degree'] == "Lower second"
df_low_sec = df[mask]
df_low_sec

mask = df['Classification of First Degree'] == "Third/Pass"
df_third = df[mask]
df_third
```
