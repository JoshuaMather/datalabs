loop through columns for col in df.columns
check first 2 characters and replace
check last 2 characters and replace
rename column if change has been made

```python
for col in df.columns:
    string = ""
    if col.startswith('01'):
        string += "G"
    elif col.startswith('02'):
        string += "S"
    elif col.startswith('03'):
        string += "B"
        
    if col.endswith('!'):
        string += "S"
    elif col.endswith('!.1'):
        string += "W"
    elif col.endswith('!.2'):
        string += "T"
    
    if string != "":
        df.rename(index=str, columns={col:string}, inplace=True)
df
        
```
